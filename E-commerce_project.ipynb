```python
# В период с января по декабрь выявите когорту с самым высоким retention на 3-й месяц. 
# Описание подхода можно найти тут. Для визуализации когортной таблицы рекомендуем 
# использовать пример из 8-го урока python, 
# раздел “Стильный урок”, степ 5. 
```

```python
# Что мне нужно для когортного анализа?
# Перед проведением когортного анализа определяю четыре параметра:
# Признак формирования когорты — ключевое действие (в данном случае - совершение пользователем покупки)
# Размер когорты — временной интервал (необходимо проанализировать покупки в период с января по декабрь с retention на 3й месяц)
# Отчетный период — 2017-01-01 по 2017-12-31 + 3 месяца (так как нам нужен retention на 3й месяц)
# Анализируемый ключевой показатель - в данном случае Retention Rate 

#  https://www.activestate.com/blog/cohort-analysis-with-python/ справочно использовал.
```

```python
# Для начала задаю даты отчетного периода
start_date = '2017-01-01'
end_date = '2018-04-01'
```

```python
period_for_cohort_df = common_df
```

```python
# Оставим только заказы, которые считаем покупкой (по аналогии выше) в нужные нам даты
period_for_cohort_df = period_for_cohort_df.query('order_approved_at >= @start_date and order_approved_at <= @end_date')
period_for_cohort_df = period_for_cohort_df.loc[(period_for_cohort_df.order_status != 'unavailable') & (period_for_cohort_df.order_status != 'canceled')] 
```

```python
# Создадим новую клонку с датой в формате "год-месяц"
period_for_cohort_df['order_date'] = period_for_cohort_df.order_purchase_timestamp.dt.strftime('%Y-%m')
```

```python
# Оставим только нужные нам колонки

period_for_cohort_df= period_for_cohort_df[['customer_unique_id','order_date']]
period_for_cohort_df
```

```python
# Найдем даты первой покупки для всех пользователей
first_order_df = period_for_cohort_df.groupby('customer_unique_id',as_index=False) \
    .agg({'order_date':'min'}) \
    .rename(columns={'order_date': 'first_order'}) \
    .sort_values(by = 'first_order', ascending = False)
first_order_df.head()
```

```python
# Соеденим данные первой покупки с данными искомого периода для когортного анализа
cohort_data_df = period_for_cohort_df.merge(first_order_df, on='customer_unique_id')
cohort_data_df.head()
```

```python
# определим тип данных

cohort_data_df.dtypes
```

```python
# Приведем даты в тип данных "дата"
cohort_data_df['first_order'] = pd.to_datetime(cohort_data_df['first_order'])
cohort_data_df['order_date'] = pd.to_datetime(cohort_data_df['order_date'])
```

```python
cohort_data_df.dtypes
```

```python
# Создаем функцию для нумерации месяцев и применяем ее к нашим объединенным данным

def diff_month(x):

    d1 = x['order_date']

    d2 = x['first_order']

    return ((d1.year - d2.year) * 12 + d1.month - d2.month)


cohort_data_df['cohort_idx'] = cohort_data_df.apply(lambda x: diff_month(x), axis=1)

cohort_data_df.head()
```

```python
# Создаем функцию для подсчета Retention Rate

def get_cohort_matrix(data, var='customer_unique_id', fun=pd.Series.nunique):

    cd = data.groupby(['first_order', 'cohort_idx'])[var].apply(fun).reset_index()

    cc = cd.pivot_table(index = 'first_order',

                        columns = 'cohort_idx',

                        values = var)

    cs = cc.iloc[:,0]

    retention = cc.divide(cs, axis = 0)

    retention = retention.round(4)

    return retention
```

```python
retention = get_cohort_matrix(cohort_data_df)
retention
```

```python
# Визуализируем данные при помощи heatmap
fig, ax = plt.subplots(figsize=(15, 5))
sns.heatmap(retention, mask=retention.isnull(), annot=True, fmt='.2%', vmin=0.0, vmax=0.01)
plt.title('Retention Rate')
plt.ylabel('Дата первой покупки (когорта)')
plt.xlabel('Номер месяца')
```

```python
# Ответ: на графике видно, что в третий месяц самый высокий retention имеет когорта 2017-05-01 (0,40%)
```

```python
# Часто для качественного анализа аудитории использую подходы, основанные на сегментации. 
# Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. 
# В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты,
#   F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. 
#  Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, 
#  frequency и monetary для интерпретации этих кластеров. 
#  Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency 
#  от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю. 
# Описание подхода можно найти тут. (23 балла)
```

```python
#  R - время от последней покупки пользователя до текущей даты 
# F - суммарное количество покупок у пользователя за всё время 
# M - сумма покупок за всё время

# методика  https://guillaume-martin.github.io/rfm-segmentation-with-python.html
```

```python
# Объединим вссе данные
full_data_df = unique_customers_ids.merge(orders_data, on='customer_id').merge(orders_items, on='order_id')
```

```python
# За настоящую дату примем дату последней покупки
current_date = full_data_df.order_approved_at.max()
current_date
```

```python
# Теперь сгруппируем данные и посчитаем R, M, F
rfm_df = full_data_df.groupby('customer_unique_id', as_index=False) \
    .agg({'order_approved_at': 'max',
        'order_id': 'count',
        'total_price': 'sum'}) \
    .rename(columns={'order_approved_at': 'recency',
                     'order_id': 'frequency',
                     'total_price': 'monetary'})
```

```python
# При помощи describe получим подробные данные, на основе которых выберем границы кластеров
rfm_df.describe()
```

```python
# Для R установим границы по 25, 50 и 75 процентилю. Чем меньше дней, тем активнее покупатель и тем выше будет его ранг.
r_q25 = rfm_df.recency.quantile(0.25)
r_q50 = rfm_df.recency.quantile(0.5)
r_q75 = rfm_df.recency.quantile(0.75)

def r_score(column):
    if column < r_q25: 
        return 4 # с последней покупки прошло менее 118 дней
    elif column >= r_q25 and column < r_q50:
        return 3 # c последней покупки прошло от 118 до 223 дней
    elif column >= r_q50 and column < r_q75:
        return 2 # c посленей покупки прошло от 223 дней до 351 дня
    elif column >= r_q75:
        return 1 # c посленей покупки прошло более 351 дня
```

```python
rfm_df['R'] = rfm_df.recency.apply(lambda x: r_score(x))
rfm_df.head()
```

```python
# Посмотрим соотношение частоты покупок срди аших клиентов
rfm_df.frequency.value_counts()
```

```python
# Большинство клиентов совершило всего 1 покупку, поэтому возьмем 1, как нижнюю границу рангов
def f_score(column):
    if column == 1:
        return 1
    elif column == 2:
        return 2
    elif column == 3:
        return 3
    elif column > 3:
        return 4
```

```python
rfm_df['F'] = rfm_df.frequency.apply(lambda x: f_score(x))
rfm_df.head()
```

```python
# Попробуем взять значения больше 75ого процентиля за вторую границу наших рангов (первая будет меньше 75%).
# Третья граница будет равняться 75ому процентилю от второй границе, а четвертая от 75ому процентилю от третьей.

m_q75_1 = rfm_df.monetary.quantile(0.75)
m_q75_2 = rfm_df.query('monetary > @m_q75_1')['monetary'].quantile(0.75)
m_q75_3 = rfm_df.query('monetary > @m_q75_2')['monetary'].quantile(0.75)
```

```python
# Получаем следующие границы рангов для M
def m_score(column):
    if column < m_q75_1:
        return 1
    elif column >= m_q75_1 and column < m_q75_2:
        return 2
    elif column >= m_q75_2 and column < m_q75_3:
        return 3
    elif column >= m_q75_3:
        return 4
```

```python
rfm_df['M'] = rfm_df.monetary.apply(lambda x: m_score(x))
rfm_df.head()
```

```python
# Соединяем оценки в один общий RFM-score
rfm_df['RFM_score'] = rfm_df['R'].map(str) + rfm_df['F'].map(str) + rfm_df['M'].map(str)
rfm_df.head()
```

```python
# RFM-score показал нам 58 возможных сегментов покупателей. В таком количестве сегментов нет особого cмысла. 
# Для простоты сначала выделим сегменты покупателей по признакам R/F:

# 1/1-2: Спящие (последняя покупка была очень давно и на маленькое количество товаров)

# 2/1-2: Cкоро заснут (покупка была давно и не на большое количество товаров)

# 2/3-4: : В зоне риска (объем покупок выше среднего, но последняя была довольно давно)

# 3/3-4: : Лояльные (покупка была недавно и купили много)

# 4/3-4: : Чемпионы (покупают часто и много)

# 3-4/2: Потенциально лояльные (покупают часто, среднее количество товара)

# 4/1: Новички (купили совсем недавно, но небольшое количество товаров. Присматриваются к нам)

# 3/1: Недавние (купили недавно, но небольшое количество товаров. Возможно не заинтересовались)

# 1/3-4: Ранее были лояльны (последний раз купили давно, но раньше покупали покупали много)
```

```python
# Добавим колонку "сегмент" к нашим данным
segt_map = {
    r'1[1-2]': 'Спящий',
    r'2[1-2]': 'Скоро уснет',
    r'2[3-4]': 'В зоне риска',
    r'3[3-4]': 'Лояльный',
    r'4[3-4]': 'Чемпион',
    r'[3-4]2': 'Потенциально лояльный',
    r'41': 'Новичек',
    r'31': 'Недавний',
    r'1[3-4]': 'Бывший лояльный'
}

rfm_df['segment'] = rfm_df['R'].map(str) + rfm_df['F'].map(str)
rfm_df['segment'] = rfm_df['segment'].replace(segt_map, regex=True)
rfm_df.head()
```

```python
# Теперь, когда мы интерпретировали сегменты, мы можем построить график, чтобы визуализировать наших покупателей по количеству
sns.set(rc={'figure.figsize':(20, 15)})
sns.barplot(data=segments_quantity, x='index', y='RFM_score')
plt.xticks(rotation=90)
plt.title('Распределение RFM сегментов по количеству покупателей')
plt.xlabel('RFM сегмент')
plt.ylabel('Количество покупателей')
```
